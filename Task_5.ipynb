```python
import pandas as pd
```

```python
data = pd.read_csv("takehome_users.csv",encoding ='ISO-8859-1')
data
```

```python
data1 = pd.read_csv("takehome_user_engagement.csv")
data1
```

```python
data.isnull().sum()
```

```python
data['last_session_creation_time'] = pd.to_datetime(data['last_session_creation_time'],unit='s')
```

```python
d = data1.copy()
d.info()
```

```python
from datetime import datetime, timedelta
d['date']=pd.to_datetime(d.time_stamp).dt.date
d['date']=pd.to_datetime(d['date'])
d.info()
```

```python
def visit_count(grp):
  return grp.rolling("7D",on='date')['user_id'].count()
d['visit_c'] = d.groupby('user_id',as_index=False,group_keys=False).apply(visit_count)
```

```python
d[d.visit_c>=3.0]
```

```python
adopt_users = d.groupby('user_id')['visit_c'].max().reset_index()
adopt_users
```

```python
adopt_users['adopted'] = adopt_users['visit_c'].apply(lambda x: 1 if x>=3 else 0)
```

```python
adopt_users.head(10)
```

```python
adopt_users['adopted'].value_counts()
```

```python
adopt_users.drop('visit_c',axis=1,inplace=True)
adopt_users.rename(columns={'user_id':'object_id'},inplace=True)

adopt_users
```

```python
user = data.join(adopt_users,on='object_id',how='left')
```
